#1.Numpy数组对象

```python
# Numpy数组通常是由相同种类的元素组成的，即数组中的数据项的类型一致。
```

```python
def pySum():
    a = list(range(10000))
    b = list(range(10000))
    c = []
    for i in range(len(a)):
        c.append(a[i]**2 + b[i]**2)
    return c
```

```python
%timeit pySum()
```

```python
import numpy as np
def npSum():
    a = np.arange(10000)
    b = np.arange(10000)
    c = a**2 + b**2
    return c
```

```python
%timeit npSum()
```

```python
#从上面的运行结果可以看出，numpy的向量化运算的效率要远远高于python的循环遍历运算（效率相差好几百倍）。 
#（1ms=1000µs）
```
